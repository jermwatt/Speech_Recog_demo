# Record audio samples and visualize raw signals

```python
# record small snippet of audio - input filename should be .wav
def record_audio(filename):
    # audio recording parameters
    FORMAT = pyaudio.paInt16
    CHANNELS = 1
    RATE = 44100
    CHUNK = 1024
    RECORD_SECONDS = 1.1
    WAVE_OUTPUT_FILENAME = filename
    
    # grab pyaudio instance
    audio = pyaudio.PyAudio()

    # start Recording
    stream = audio.open(format=FORMAT, channels=CHANNELS,
                    rate=RATE, input=True,
                    frames_per_buffer=CHUNK)
    print "recording..."
    frames = []

    for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(data)
    print "finished recording"

    # stop Recording
    stream.stop_stream()
    stream.close()
    audio.terminate()

    waveFile = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
    waveFile.setnchannels(CHANNELS)
    waveFile.setsampwidth(audio.get_sample_size(FORMAT))
    waveFile.setframerate(RATE)
    waveFile.writeframes(b''.join(frames))
    waveFile.close()
```

```python
# record audio samples - here I've recorded 3 samples (2 versions of the word 'dog' and one 'cat')
record_audio('audio_files/dog1.wav')
```

```python
# short function to plot wav audio file - raw signal
def plot_audio_signal(filename,subplot_title,ax):
    # load in audio file
    spf = wave.open(filename,'r')

    # extract audio from wav file
    sig = spf.readframes(-1)
    sig = np.fromstring(sig, 'Int16')
    fs = spf.getframerate()
    Time=np.linspace(0, len(sig)/fs, num=len(sig))

    # set up figure and plot
    ax.set_title(subplot_title)
    ax.plot(Time,sig,color = 'r')
    ax.set_yticks([],[])
    ax.axis('off')  
```

Lets plot one audio signal to see what it looks like

```python
# create subplot figure
f, (ax1) = plt.subplots(1, 1, sharey=True,figsize=(10, 4))
plot_audio_signal('audio_files/dog1.wav','dog',ax1)
```

# How can we distinguish simple prototype signals?

Audio signals - as interpreted by the computer (and your ear) - are complicated 2-dimensional wave forms.  Remember: our goal here is to understand how we could possibly classify such signals.  

A simplified classification scenario: If we have the signal on the left above in our data base, and the two signals in the middle and right are new, how can we tell which is more similar to our original signal?  

Lets simplify the situation, and first understand how we could possibly do this with simple prototype audio signals.

## Fact 1: complex signals can be decomposed using simple waves
This is a very old concept - several centuries old in fact - that complex signals (like an audio signal) can be decomposed or represented using simple waves.  

This gif [from Wikipedia](https://en.wikipedia.org/wiki/Fourier_transform) shows the entire process of representing an audio signal as the **histogram** of weights on its sine wave representation (ignore the formulae).

<img src="images/fourier_transform.gif\" width=400 height=400/>

In other words, we can represent an audio signal by the weights on its sine wave decomposition.

<img src="images/fourier_rep.png\" width=600 height=600/>
